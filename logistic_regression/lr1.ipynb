```python
x = np.linspace(-10,10,100,1000)
sig = 1/(1 + np.exp(-x))
plt.figure(figsize=(10, 5))
plt.grid()
plt.plot(x, sig, "r-", linewidth=2, label=r"$\sigma(x) = \frac{1}{1 + e^{-x}}$")
plt.xlabel("x")
plt.legend(fontsize=15)
plt.show()
```
